```python
# 매출
매출_2024 = pd.read_csv('/content/drive/MyDrive/DataMining2025/행정동_추정매출_2024.csv')
# 1. 수치형 변수만 선택 (기준_년분기_코드, 행정동_코드 제외)
수치형_변수 = 매출_2024.select_dtypes(include='number').columns.tolist()
수치형_변수 = [col for col in 수치형_변수 if col not in ['기준_년분기_코드', '행정동_코드']]

# 2. 기준_년분기_코드, 행정동_코드 기준 중복 여부 확인
중복_키 = 매출_2024.duplicated(subset=['기준_년분기_코드', '행정동_코드'], keep=False)

# 3. 중복된 경우만 필터링
중복_데이터 = 매출_2024[중복_키]

# 4. 기준_년분기_코드, 행정동_코드 기준으로 수치형 변수 합산
중복_집계 = 중복_데이터.groupby(['기준_년분기_코드', '행정동_코드'])[수치형_변수].sum().reset_index()

원하는_컬럼 = [
    '기준_년분기_코드', '행정동_코드',
    '당월_매출_금액', '주중_매출_금액', '주말_매출_금액',
    '남성_매출_금액', '여성_매출_금액',
    '연령대_10_매출_금액', '연령대_20_매출_금액', '연령대_30_매출_금액',
    '연령대_40_매출_금액', '연령대_50_매출_금액', '연령대_60_이상_매출_금액'
]

# 위에서 만든 중복 기준 집계 데이터에서 원하는 컬럼만 선택
중복_집계_필터 = 중복_집계[원하는_컬럼]

# 결과 출력
print(중복_집계_필터.shape)
중복_집계_필터.head()
```

```python
# 파일 불러오기
매출 = pd.read_csv('/content/drive/MyDrive/DataMining2025/매출전체_전처리.csv', encoding='utf-8-sig')
유동인구 = pd.read_csv('/content/drive/MyDrive/DataMining2025/행정동_유동인구_전처리.csv', encoding='utf-8-sig')
상주인구 = pd.read_csv('/content/drive/MyDrive/DataMining2025/상주인구_전처리.csv', encoding='utf-8-sig')
시설 = pd.read_csv('/content/drive/MyDrive/DataMining2025/시설_전처리_확장.csv', encoding='utf-8-sig')
아파트 = pd.read_csv('/content/drive/MyDrive/DataMining2025/아파트_가격_전처리.csv', encoding='utf-8-sig')
소득 = pd.read_csv('/content/drive/MyDrive/DataMining2025/소득_전처리.csv', encoding='utf-8-sig')

# 기준 키
key = ['기준_년분기_코드', '행정동_코드']

# 중복 컬럼 제거
상주인구=상주인구.drop(columns=['행정동_코드_명'])
시설=시설.drop(columns=['행정동_코드_명'])
아파트=아파트.drop(columns=['행정동_코드_명'])
소득=소득.drop(columns=['행정동_코드_명'])

# 병합
통합 = 매출.merge(유동인구, on=key, how='left').merge(상주인구, on=key, how='left').merge(시설, on=key, how='left').merge(아파트, on=key, how='left').merge(소득, on=key, how='left')

통합['비교통_집객시설_수'] = 통합['집객시설_수'] - 통합['대중교통_수']

# 결과 확인
print(통합.shape)
print(통합.isnull().sum())
통합.head()
```
