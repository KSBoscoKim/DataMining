```python
import json
```

```python
geo_data = json.load(open('HangJeongDong_ver20241231.geojson',encoding='utf-8'))
```

```python
geo_data
```

```python
import pandas as pd
```

```python
file = 'final.csv'
data = pd.read_csv(file)
```

```python
data['행정동_코드'] = data['행정동_코드'].astype(int)*100
data_use = data.loc[:,:]
data_use
```

```python
folium.GeoJson(geo_data,name='행정동_코드')
```

```python
# 클러스터 분포
seoul = folium.Map(location=[37.5536067,126.9674308], zoom_start=11)
choropleth = folium.Choropleth(
geo_data=geo_data,
name=data_use,
data=data_use,
columns=('행정동_코드','계층_클러스터'),
key_on='feature.properties.adm_cd2',
fill_color='YlGnBu',
fill_opacity=0.7,
line_opacity=0.5,
legend_name=' 클러스터 분포',
nan_fill_color='white',
nan_fill_opacity=1).add_to(seoul)
seoul
```
